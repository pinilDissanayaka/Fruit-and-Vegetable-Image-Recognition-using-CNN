```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, BatchNormalization, MaxPooling2D, Dropout, Flatten, Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping


import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
traindir = '/content/train'
testdir = '/content/test'
validdir = '/content/validation'
```

```python
IMG_SIZE = [300, 300, 3]
outsize = len(os.listdir(traindir))

print(f"Input size = {IMG_SIZE}")
print(f"Output size = {outsize}")
```

```python
dir = [traindir, testdir, validdir]

for i in dir:
  print(f"\nFor {i} sample has : ")
  for j in range(len(os.listdir(i))):
    currdir = i + '/' + os.listdir(i)[j]
    print(f"{currdir} has {len(os.listdir(currdir))}")
```

```python
traindatagen = ImageDataGenerator(
    rescale=1./255,
    featurewise_center=True,
    featurewise_std_normalization=True,
    rotation_range=20,
    width_shift_range=0.3,
    height_shift_range=0.9,
    horizontal_flip=True)

testdatagen = ImageDataGenerator(
    rescale=1./255)

validdatagen = ImageDataGenerator(
    rescale=1./255)
```

```python
traindf = traindatagen.flow_from_directory(
        traindir,
        target_size=(300, 300),
        batch_size=32,
        class_mode='categorical')

testdf = traindatagen.flow_from_directory(
        testdir,
        target_size=(300, 300),
        batch_size=32,
        class_mode='categorical')


validdf = traindatagen.flow_from_directory(
        validdir,
        target_size=(300, 300),
        batch_size=32,
        class_mode='categorical')
```

```python
model = Sequential()

model.add(Conv2D(32, (3 * 3), padding='same', input_shape=IMG_SIZE))
model.add(BatchNormalization())
model.add(MaxPooling2D(2, padding='same'))
model.add(Conv2D(32, (3 * 3), padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D(2, padding='same'))
model.add(Dropout(0.2))
model.add(Conv2D(64, (3 * 3), padding='same'))
model.add(BatchNormalization())
#model.add(Activation('relu'))
model.add(MaxPooling2D(2, padding='same'))
model.add(Dropout(0.3))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(units=256, activation='relu'))
model.add(Dense(units=256, activation='relu'))
model.add(Dense(units=outsize, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer="Adam",
              loss="categorical_crossentropy",
              metrics=['acc'])
```

```python
earlystopping = EarlyStopping(monitor = 'val_loss',
                              patience = 1,
                              verbose = 1)
```

```python
model_hist = model.fit_generator(
    traindf,
    epochs = 10,
    verbose = 1,
    validation_data = validdf,
    steps_per_epoch = len(traindf),
    validation_steps = len(validdf)
)
```
